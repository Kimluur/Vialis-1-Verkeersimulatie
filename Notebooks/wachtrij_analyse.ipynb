## Wachtrij-data analyse

In deze notebook gaan we de door ons gegenereerde wachtrij-data analyseren. 
Dit doen we voor alle data die we hebben en specifiek voor 8 & 10 januari.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import random
```

```python
df = pd.read_csv('Wachtrij.csv')
df.head()
```

```python
def get_perHour(df):
    multi = 36000

    perHour = df.groupby(np.arange(len(df))//multi).mean()

    every_index = []
    [every_index.append(df.iloc[index*multi]['time']) for index, row in perHour.iterrows()]

    perHour['time'] = every_index
    perHour = perHour.set_index('time')
    
    return perHour
```

## Alle dagen

```python
# Gemiddelde wachtrij per uur

all_per_hour = get_perHour(df)    
lines = all_per_hour.plot.line()
plt.xticks(rotation=70)
plt.title('Gemiddelde wachtrij per uur')
plt.xlabel('DateTime')
plt.ylabel('Gemiddelde wachtrij')    
```

```python
sns.heatmap(df[["01", "03", "04","05", "11", "12", "41"]], yticklabels=False, cmap="mako", vmin=1)
```

## 8 Januari

```python
# Gemiddelde wachtrij per uur (8 Januari)

df_8_1 = df[(df["time"]>= "08-01-2021 00:00:00.0") & (df["time"] <= "08-01-2021 23:59:59.9")]


all_per_hour = get_perHour(df_8_1)    
lines = all_per_hour.plot.line()
plt.xticks(rotation=70)
plt.title('Gemiddelde wachtrij per uur (8 Januari)')
plt.xlabel('DateTime')
plt.ylabel('Gemiddelde wachtrij') 
```

```python
sns.heatmap(df_8_1[["01", "03", "04","05", "11", "12", "41"]], yticklabels=False, cmap="mako", vmin=1)
```

## 10 Januari


```python
# Gemiddelde wachtrij per uur (10 Januari)

df_1_10 = df[(df["time"]>= "10-01-2021 00:00:00.0") & (df["time"] <= "10-01-2021 23:59:28.0")]


all_per_hour = get_perHour(df_1_10)    
lines = all_per_hour.plot.line()
plt.xticks(rotation=70)
plt.title('Gemiddelde wachtrij per uur (10 Januari)')
plt.xlabel('DateTime')
plt.ylabel('Gemiddelde wachtrij') 
```

```python
sns.heatmap(df_1_10[["01", "03", "04","05", "11", "12", "41"]], yticklabels=False, cmap="mako", vmin=1)
```
